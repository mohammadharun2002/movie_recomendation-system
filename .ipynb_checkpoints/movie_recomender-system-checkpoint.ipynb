```python
import pandas as pd
import numpy as np
import ast
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv");
credits = pd.read_csv("tmdb_5000_credits.csv");
```

```python
# genres, id, keywords, title, overview, cast, crew

movies = movies.merge(credits, on='title')

movies = movies[['genres', 'id', 'keywords', 'title', 'overview', 'cast', 'crew']]
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].keywords
```

```python
# '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
# ["Action", "Adventure", "Fantasy", "Science Fiction"]
```

```python
def convertToList(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i["name"])
    return L
```

```python
movies["genres"] = movies["genres"].apply(convertToList)
```

```python
movies["genres"].head()
```

```python
movies["keywords"] = movies["keywords"].apply(convertToList)
```

```python
movies["keywords"].head()
```

```python
def convertToList3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i["character"])
            counter += 1
        else:
            break;
    return L
```

```python
movies["cast"] = movies["cast"].apply(convertToList3)
```

```python
movies["cast"].head()
```

```python
def GetDirector(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i["job"] == "Director":
            L.append(i["name"])
            break;
    return L
```

```python
movies["crew"] = movies["crew"].apply(GetDirector)
```

```python
movies.head()
```

```python
movies["overview"] = movies["overview"].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies["genres"] = movies["genres"].apply(lambda x:[i.replace(" ", "") for i in x])
movies["keywords"] = movies["keywords"].apply(lambda x:[i.replace(" ", "") for i in x])
movies["cast"] = movies["cast"].apply(lambda x:[i.replace(" ", "") for i in x])
movies["crew"] = movies["crew"].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
movies["tags"] = movies["overview"] + movies["keywords"] + movies["cast"] + movies["crew"] +  movies["genres"]
```

```python
new_df = movies[["id", "title", "tags"]]
```

```python
new_df.head()
```

```python
new_df["tags"] = new_df["tags"].apply(lambda x:" ".join(x))
```

```python
new_df["tags"][0]
```

```python
new_df.head()
```

```python
new_df["tags"] = new_df["tags"].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
def Stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y) 
```

```python
new_df["tags"] = new_df["tags"].apply(Stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words="english")
```

```python
vectors = cv.fit_transform(new_df["tags"]).toarray()
```

```python
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors).shape
```
